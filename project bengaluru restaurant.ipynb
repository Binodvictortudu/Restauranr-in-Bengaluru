```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

```

```python
#reading csv file
```

```python
df=pd.read_csv('zomato.csv')
```

```python
#To see first 5 rows of table
```

```python
df.head()
```

```python
#To see no.of rows and columns and to know which column contain null values
```

```python
df.info()
```

```python
#To list all the columns
```

```python
df.columns
```

```python
#To remove some column which we don't want
```

```python
df=df.drop(['url','phone','dish_liked','menu_item','reviews_list'], axis=1)
```

```python
#To display top 5 rows aftr removing some column
```

```python
df.head()
```

```python
#To permanently remove the duplicate data
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
#To display all unique value from rate column
```

```python
df['rate'].unique()
```

```python
#To remove 'NEW','-','/5'
```

```python
def handelrate(value):
  if(value=='NEW' or value=='-'):
        return np.nan
  else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
    
df['rate']=df['rate'].apply(handelrate)
df.head()
```

```python
#Total null value in rate column
```

```python
df['rate'].isnull().sum()
```

```python
#filling null values with the mean
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.info()
```

```python
#to rename column name
```

```python
df.rename(columns={'approx_cost(for two people)':'cost2plate','listed_in(type)':'Type'},inplace=True)
df.head()
```

```python
#to display all unique location
```

```python
df['location'].unique()
```

```python
#dropping two column
```

```python
df=df.drop(['listed_in(city)','menu_item'], axis=1)
```

```python
df.head()
```

```python
df['cost2plate'].unique()
```

```python
#to remove ','between price which prevents them from becoming integer
```

```python
def handlecoma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
    
df['cost2plate']=df['cost2plate'].apply(handlecoma)
```

```python
df['cost2plate'].unique()
```

```python
df.head()
```

```python
#to count diff type of restaurant
#if same type of restaurant is less than 1000,it will come under 'others'
```

```python
rest_types=df['rest_type'].value_counts()
```

```python
rest_types_lessthan1000=rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
    
df['rest_type']=df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
#to count no. of restaurant in any particular location
#if no. of restaurant is less than 1000 on any particlar location,it will be considered as others 'others'
```

```python
location=df['location'].value_counts()
```

```python
location_lessthan300=location[location<300]
location_lessthan300
```

```python
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
    
df['location']=df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
cuisines=df['cuisines'].value_counts()
```

```python
cuisines_lessthan100=cuisines[cuisines<100]
```

```python
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
    
df['cuisines']=df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```

```python
df['Type'].value_counts()
```

```python
address=df['address'].value_counts()
```

```python
address_lessthan50=address[address<50]
```

```python
def handle_address(value):
    if(value in address_lessthan50):
        return 'others'
    else:
        return value
    
df['address']=df['address'].apply(handle_address)
df['address'].value_counts()
```

```python
df.head()
```
